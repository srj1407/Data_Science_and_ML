Stochastic Gradient Descent

```python
def train(x, y, m, c, eta):
  y_pred = m*x+c
  dm = 2*(y_pred-y)*x
  dc = 2*(y_pred-y)
  m = m-eta*dm
  c = c-eta*dc
  error = ((y_pred-y)**2)
  return m, c, error
```

```python
def one_loop_random(l, t_sq, m, c, eta):
  random_idx = np.arange(len(l))
  np.random.shuffle(random_idx)
  for idx in random_idx:
    m, c, error = train(l[idx], t_sq[idx], m, c, eta)
  return m, c, error
```

```python
def train_stochastic(l, t_sq, m, c, eta, iterations = 1000):
  for i in range(iterations):
    m, c, error = one_loop_random(l, t_sq, m, c, eta)
  return m, c, error
```

```python
m, c=0, 0
lr=0.01
for i in range(5):
  m, c, error = train_stochastic(l, t_sq, m, c, lr, iterations = 10)
  y=m*l+c
plt.plot(l, t_sq, '.k');
plt.plot(l, y);
plt.xlabel("$L$", fontsize=15)
plt.ylabel("$T^2$", fontsize=15)
plt.show()
```

```python
m, c=0, 0
lr=0.01
errs=[]
for i in range(10):
  m, c, error = train_stochastic(l, t_sq, m, c, lr, iterations = 10)
  errs.append(error)
  y=m*l+c
plt.plot(range(10), errs);
plt.xlabel("Iteration", fontsize=12)
plt.ylabel("Error", fontsize=12)
plt.show()
```

```python
def train(x, y, m, c, eta):
  y_pred = m*x+c
  err = (y_pred-y)**2
  const = 2/len(x)
  dm=const*np.sum((y_pred-y)*x)
  dc=const*np.sum(y_pred-y)
  m=m-eta*dm
  c=c-eta*dc
  return m, c, err
```

```python
def minibatch(x, y, m, c, eta, batch_size = 10):
  for i in range(0, len(x), batch_size):
    m, c, err = train(x[i:min(len(x), i+batch_size)], y[i:min(len(x), i+batch_size)], m, c, eta)
  return m, c, err
```

```python
def full_gd(l, t_sq, batch_size, eta, iterations = 100):
  m, c=0, 0
  for i in range(iterations):
    m, c, err = minibatch(l, t_sq, m, c, eta, batch_size)
  return m, c, err
```

```python
m, c, err=full_gd(l, t_sq, 10, 0.01, 1000)
y=m*l+c
plt.plot(l, t_sq, '.k');
plt.plot(l, y);
plt.xlabel("$L$", fontsize=15)
plt.ylabel("$T^2$", fontsize=15)
plt.show()
```
